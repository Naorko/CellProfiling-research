```python
import numpy as np
import pandas as pd
from glob import glob
```

```python
indexes = glob('/storage/users/g-and-n/plates/tabular_indexes/*')
```

```python
idx_dfs = [pd.read_csv(ind).groupby(['Plate','Metadata_broad_sample', 'Image_Metadata_Well']).count()['ObjectNumber'] for ind in indexes]
```

```python
idx_dfs = [idx_df.reset_index() for idx_df in idx_dfs]
```

```python
platemaps = pd.concat(idx_dfs).groupby('Metadata_broad_sample').agg(lambda l: ','.join([str(x)for x in l]))
```

```python
matches = platemaps['Plate'].unique()
```

```python
uniq_mch = np.unique([set(m.split(',')) for m in matches if len(m.split(','))<100])
```

```python
uniq, cnts = np.unique([len(m) for m in uniq_mch], return_counts=True)
```

```python
plt_n = [len(set.union(*[u for u in uniq_mch if len(u)==s])) for s in uniq]
```

```python
plt_n_lte = [len(set.union(*[u for u in uniq_mch if len(u)<=s])) for s in uniq]
plt_n_gte = [len(set.union(*[u for u in uniq_mch if len(u)>=s])) for s in uniq]
```

```python
all_p=set.union(*uniq_mch)
```

```python
plt_n_lte_c = [len(all_p-set.union(*[u for u in uniq_mch if len(u)<=s])) for s in uniq]
plt_n_gte_c = [len(all_p-set.union(*[u for u in uniq_mch if len(u)>=s])) for s in uniq]
```

```python
plt_n_lte_cp = [dict(zip(*np.unique(
    [len(ut-set.union(*[u for u in uniq_mch if len(u)<=s])) for ut in uniq_mch if len(ut)>s],
    return_counts=True))) 
                     for s in uniq]
plt_n_gte_cp = [dict(zip(*np.unique(
    [len(ut-set.union(*[u for u in uniq_mch if len(u)>=s])) for ut in uniq_mch if len(ut)<s],
    return_counts=True))) 
                     for s in uniq]
```

```python
res = pd.DataFrame({'Size':uniq,
                    'Count':cnts,
                    'Unique Plates':plt_n,
                    'Aggragate <=s':zip(plt_n_lte,plt_n_lte_c,plt_n_lte_cp),
                    'Aggragate >=s':zip(plt_n_gte,plt_n_gte_c,plt_n_gte_cp)})
res.set_index('Size')
```

# Tabular Metadata

```python
import pandas as pd
import random
import os
from tqdm import tqdm
```

```python
split_field = 'Image_Metadata_Well'
mt_df = pd.read_csv('/storage/users/g-and-n/plates/tabular_metadata.csv', dtype={'Plate': int, 'Count': int})
mt_df[split_field] = mt_df[split_field].apply(eval)
```

```python
mt_df
```

```python
by_fld = 'Image_Metadata_Well'
for idx, row in tqdm(mt_df[mt_df.Mode == 'train'][mt_df[mt_df.Mode == 'train'].Image_Metadata_Well.apply(len) == 16].iterrows()):
    p, t, mode, wells, _ = row

    val_wells = set(random.sample(wells, 4))
    train_wells = {w for w in wells if w not in val_wells}
    
    df = pd.read_csv(os.path.join('/storage/users/g-and-n/plates/csvs_processed', f'{p}_{lbl}_train.csv'))
    is_train = df[by_fld].isin(train_wells)
    is_val = df[by_fld].isin(val_wells)
    train_count = is_train.sum()
    val_count = is_val.sum()
    df[is_val].to_csv(os.path.join('/storage/users/g-and-n/plates/csvs_processed', f'{p}_{lbl}_val.csv'), index=False)
    df[is_train].to_csv(os.path.join('/storage/users/g-and-n/plates/csvs_processed', f'{p}_{lbl}_train.csv'), index=False)

    mt_df.at[idx, by_fld] = train_wells
    mt_df.at[idx, 'Count'] = train_count 

    mt_df = mt_df.append({'Plate':p,
                  by_fld: lbl,
                  'Mode': 'val',
                  split_field: val_wells,
                  'Count':val_count}, ignore_index=True)
    
mt_df.sort_values(['Plate', 'Metadata_ASSAY_WELL_ROLE', 'Mode']).to_csv('/storage/users/g-and-n/plates/tabular_metadata.csv', index=False)
```
